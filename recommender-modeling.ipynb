```python
use_col = ['bracelet', 'ring', 'necklace',
       'earring', 'choker', 'pendant', 'mesh', 'gold', 'silver', 'white',
       'titanium', 'ip', 'rose', 'steel', 'oxidized', 'pearl',
       'chalcedony', 'rhodonite', 'opal', 'beryllium', 'morganite', 'citrine',
       'topaz', 'praseolite', 'carnelian', 'amethyst', 'malachite',
       'amazonite', 'labradorite', 'sapphires', 'agate', 'quartzite',
       'gemstones', 'spinel', 'onyx', 'diamonds', 'ruby']
```

```python
df_rec = df_items[use_col]
```

```python
from sklearn.preprocessing import normalize
```

```python
normed_data = normalize(df_rec)
```

```python
df_rec.head()
```

```python
from sklearn.metrics.pairwise import linear_kernel

# will calculate similarity between rows between data above, each row and columns represent id of items
cosine_sim = linear_kernel(normed_data, normed_data)
```

```python
cosine_sim.shape
```

```python
df_sim = pd.DataFrame(cosine_sim, index=df_items.ID, columns=df_items.ID)
```

```python
df_sim.head()
```

```python
def get_rec(Id):
    similar = df_sim[Id].sort_values(ascending=False).drop(Id)
    return similar.index[:5]
```

```python
get_rec(1.0)
```

```python
# These are the best related items with 1.0 

df_items.iloc[get_rec(1.0)]
```
